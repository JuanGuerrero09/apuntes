```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
import scipy.stats as st
import math
```

```python
n = 100
x = np.random.rand(n)
y = 3 + 7 * x + np.random.rand(n)
print(y)
plt.scatter(x, y)
plt.show()
```

```python
slope, intercept, rvalue, pvalue, error = st.linregress(x, y )
print(slope, intercept, rvalue, pvalue, error)
```

```python
plt.scatter(x, y)
plt.plot(x.min(), x.max(), [slope * x.min() + intercept, slope * x.max() + intercept], color='red')
```
