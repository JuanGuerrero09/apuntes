```python
# Generar los gráficos box-plot de todos los parámetros con títulos y etiquetas
boxplot_data = df
fig, axs = plt.subplots(5,2, figsize=(8,20))
titles = list(df.columns)
for i, title in enumerate(titles):
    row, col = divmod(i, 2)
    if title in data_without_nan:
        axs[row, col].boxplot(data_without_nan[title])
    else:
        axs[row, col].boxplot(boxplot_data[title])
    axs[row, col].set_title(title)
    axs[row, col].grid(axis="x", linestyle='-')
plt.show()
df.describe()
```

```python
# Generar los gráficos box-plot de todos los parámetros con títulos y etiquetas
boxplot_data = df
fig, axs = plt.subplots(5,2, figsize=(8,25))
titles = list(df.columns)
for i, title in enumerate(titles):
    row, col = divmod(i, 2)
    if title in data_without_nan:
        axs[row, col].hist(data_without_nan[title], label="Hola")
    else:
        axs[row, col].hist(boxplot_data[title])
    axs[row, col].set_title(title)
    axs[row, col].set_xlabel("Valor")
    axs[row, col].set_ylabel("Frecuencia")
    axs[row, col].grid( linestyle='-')
plt.tight_layout()
plt.show()
```

```python
# revisar https://www.studentenwerk-dresden.de/english/wohnen/faq-1.html
from rasterio import plot
```

```python
# Realizar una representación múltiple de las bandas.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 10))  
plot.show(banda4, ax=ax1,) #red  
plot.show(banda8, ax=ax2,) #nir
fig.tight_layout()
```

```python
# Plotear el NDVI.
print(banda4.read(1).shape)
print(banda8.read(1).shape)

print(banda4.crs)
print(banda8.crs)

red_band = banda4.read(1).astype(np.float64) #red band
nir_band = banda8.read(1).astype(np.float64) #near infrared
# Calculando el NDVI
ndvi = np.where((red_band + nir_band) == 0, 0, (nir_band - red_band) / (nir_band + red_band))
# Guardando el NDVI en un archivo GeoTIFF
ndviImage = rasterio.open('./Data Examen 2/Landsat8/ndviImage.tiff', 'w', driver='GTiff',
                         height=banda4.height, width=banda4.width,
                         count=1, dtype='float64', crs=banda4.crs,
                         transform=banda4.transform)
ndviImage.write(ndvi, 1)
ndviImage.close()
```

```python
ndvi = rasterio.open('./Data Examen 2/Landsat8/ndviImage.tiff')


fig = plt.figure(figsize=(8, 10))
ax = fig.add_subplot(111)
plot.show(ndvi, ax=ax, cmap='RdYlGn')


ax.set_title('Imagen NDVI')

plt.show()
```
