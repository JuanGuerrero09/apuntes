```python
total_nan = df.isna().sum().sum()
print(f"Total NaN en el DataFrame: {total_nan}")
# Contar cuántos NaN hay por cada columna
nan_por_columna = df.isna().sum()
print(nan_por_columna)
```

```python
# Debido a que hay na en Sulfate, ph y Trihalomethanes se procede a drop na en estos
sulf = df["Sulfate"].dropna()
ph = df["ph"].dropna()
trihalo = df["Trihalomethanes"].dropna()

data_without_nan = pd.DataFrame({"Sulfate": sulf, "ph": ph, "Trihalomethanes": trihalo})
data_without_nan
total_nan = data_without_nan.isna().sum().sum()
print(f"Total NaN en el DataFrame: {total_nan}")
# Contar cuántos NaN hay por cada columna
nan_por_columna = data_without_nan.isna().sum()
print(nan_por_columna)
```

```python
# Debido a que hay na en Sulfate, ph y Trihalomethanes se procede a drop na en estos
data_without_nan = df[["Sulfate", "ph", "Trihalomethanes"]].dropna()
data_without_nan
```

```python
total_nan = data_without_nan.isna().sum().sum()
print(f"Total NaN en el DataFrame: {total_nan}")
# Contar cuántos NaN hay por cada columna
nan_por_columna = data_without_nan.isna().sum()
print(nan_por_columna)
print("ph" in data_without_nan)
```

```python
# Generar los gráficos box-plot de todos los parámetros con títulos y etiquetas
boxplot_data = df
fig, axs = plt.subplots(5,2, figsize=(8,20))
titles = list(df.columns)
for i, title in enumerate(titles):
    row, col = divmod(i, 2)
    if title in data_without_nan:
        axs[row, col].boxplot(data_without_nan[title])
    else:
        axs[row, col].boxplot(boxplot_data[title])
    axs[row, col].set_title(title)
    axs[row, col].grid(axis="x", linestyle='-')
plt.show()
df.describe()
```

```python
# Generar los gráficos box-plot de todos los parámetros con títulos y etiquetas
boxplot_data = df
fig, axs = plt.subplots(5,2, figsize=(8,25))
titles = list(df.columns)
for i, title in enumerate(titles):
    row, col = divmod(i, 2)
    if title in data_without_nan:
        axs[row, col].hist(data_without_nan[title], label="Hola")
    else:
        axs[row, col].hist(boxplot_data[title])
    axs[row, col].set_title(title)
    axs[row, col].set_xlabel("Valor")
    axs[row, col].set_ylabel("Frecuencia")
    axs[row, col].grid( linestyle='-')
plt.tight_layout()
plt.show()
```
